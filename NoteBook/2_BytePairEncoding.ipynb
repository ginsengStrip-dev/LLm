```python
with open("../data/combined_text.txt", "r", encoding="utf-8") as f:
    text_sequence = f.read()

len(text_sequence)
```

```python
import sys
sys.path.append('..')
```

```python
from minbpe import BasicTokenizer

tokenizer = BasicTokenizer()
tokenizer.train(text_sequence, vocab_size=2492)
```

```python
vocab = tokenizer.vocab
len(vocab)
```

```python
tokenizer.encode("hi, how are you doing?")
```

```python
tokenizer.decode([462, 273, 400, 518, 345, 279, 63])
```

```python
max_vocab_id = list(tokenizer.vocab.keys())[-1]
tokenizer.special_tokens = {
    "<|startoftext|>": max_vocab_id + 1,
    "<|separator|>": max_vocab_id + 2,
    "<|endoftext|>": max_vocab_id + 3,
    "<|unk|>": max_vocab_id + 4,
    "<|padding|>": max_vocab_id + 5,
}
```

```python
len(tokenizer.encode(text_sequence))
```

```python
tokenizer.save(file_prefix="../Data/tokenizer/my_tokenizer")
```
